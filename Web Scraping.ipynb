### : - Libraries commonly used in web scraping include:

#####  --->   Beautiful Soup: 
         A Python library for pulling data out of HTML and XML files. It provides simple methods for navigating and searching the parse tree.

#####  --->   Requests: 
         A Python library for making HTTP requests. It allows you to send requests to web servers and handle the responses.

#####  --->   Scrapy: 
         An open-source and collaborative web crawling framework for Python. It provides a high-level API for extracting data from websites.

#####  --->   Selenium: 
         A web browser automation tool that can be used for web scraping. It allows you to interact with web pages and simulate user actions.

#####  --->   Pandas: 
        While not specifically for web scraping, Pandas is a powerful data manipulation library in Python that is often used to clean and analyze data extracted from websites.

###   ** Here are the common steps involved in web scraping:

: - Get the URL ---->  Identify the website from which you want to scrape data and obtain the URL.

: - Send an HTTP Request ----> Use a library like requests in Python to send a request to the URL and retrieve the HTML content of the webpage.

: - Parse the HTML ----> Use a library like Beautiful Soup to parse the HTML content and create a parse tree that can be used to extract data.

: - Find the Data ----> Use the methods provided by the parsing library to locate the specific data you want to extract, such as text, links, or images.

: - Extract the Data ----> Once you have located the data, extract it using the appropriate methods provided by the parsing library.

: - Store the Data ----> Store the extracted data in a format that is suitable for further analysis or use, such as a CSV file or a database.

: - Repeat for Multiple Pages ----> If you need to scrape data from multiple pages, repeat the above steps for each page, updating the URL as needed.

: - Handle Errors and Exceptions ----> Implement error handling to deal with situations where the website is unavailable or the data is not in the expected format.

: - Respect Robots.txt ----> Check the website's robots.txt file to ensure you are not violating any terms of service or scraping restrictions.

## Web Scraping Challenges and How I Tackled Them

### While web scraping, I've encountered a few challenges, such as:


1. **Finding the right data : -**   Sometimes, it's hard to locate the specific information I need on a webpage.
   
2. **Dealing with dynamic content : -**   Websites that load content dynamically can make it tricky to scrape the most up-to-date information.

3. **Pagination : -**   Scraping multiple pages of results can be challenging, as I need to navigate through different pages to get all the data I need.

4. **Anti-scraping measures : -**   Some websites implement measures like IP blocking or CAPTCHAs, making it difficult to scrape their content. 


### When faced with these challenges, I've developed strategies to overcome them:

1. **Finding the right data : -** I approach this by carefully examining the webpage structure and using tools like Chrome Developer Tools to inspect elements and identify the data locations accurately.

2. **Dealing with dynamic content : -** I look for websites that present information in a more static or easily accessible format, avoiding those with heavy dynamic content.

3. **Pagination : -** To handle pagination, I create scripts that automatically navigate through pages, extracting data from each one until all relevant information is collected.

4. **Anti-scraping measures : -** For websites with anti-scraping measures, I use custom headers in my requests to mimic a regular user, ensuring my scraping activities are not blocked.

````python
# Import necessary libraries


import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.text_rank import TextRankSummarizer
````

````python
# Step 1: Get data from our file 

with open('web_scraping_data.txt', 'r', encoding='utf-8') as file:
    # Read the lines of text data from the file
    text_data = file.readlines()
````

````python
# Step 2: Vectorization
# Convert the text data into a format suitable for machine learning 
# We do this becaues machine can underded 

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(text_data)
````

````python
# Step 3: Run Queries in vector Database

# Define a query to search for similar documents
query = "Electric vehicles (EVs) make up"

# Convert the query into a format suitable for comparison
query_vector = vectorizer.transform([query])

# Compare the query vector with the document vectors to find similar documents
cosine_similarities = cosine_similarity(X, query_vector).flatten()

# Sort the documents we will get top 4 near by outputes 

related_docs_indices = cosine_similarities.argsort()[:-5:-1]

# Get the score how much they are same 
related_docs_scores = cosine_similarities[related_docs_indices]
````

````python
# Step 4: Text Summarization
# Create blank list to save output

summaries = []


# Iterate over the indices of the top documents
for doc_index in related_docs_indices:
    
    # Create parser for the document text
    parser = PlaintextParser.from_string(text_data[doc_index], Tokenizer("english"))
    
    # Initialize TextRankSummarizer for summarization
    summarizer = TextRankSummarizer()
    
    # Summarize the document to 3 sentences
    summary = summarizer(parser.document, 3)
    
    # Convert the summary sentences into a single string and add it to the summaries list
    summary_text = " ".join([str(sentence) for sentence in summary])
    summaries.append(summary_text)
````

````python
# Step 5: To create report we need to use this dataframe fromet 

# Create a DataFrame to store the top documents, their similarity scores, and summaries

report_df = pd.DataFrame({'Document': [text_data[i] for i in related_docs_indices],
                          'Similarity Score': related_docs_scores,
                          'Summary': summaries})
````
